```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import *
# from Data_sorting import Data_formatter_cycles
```

```python
data = pd.read_excel('C:/Users/ADITYA/OneDrive - Imperial College London/Year 4/FYP/Final-year-project/Data formatting/Dataset_scaled_denoised.xlsx')
data = data[data['I'] == 1.6]
X = data.iloc[1:-1,:11]
# X = X.drop('I', axis = 1)
y = data.iloc[2:,:11]
X = X.drop('I', axis = 1)
```

```python
print(y)
```

```python
S8 = y['S8_cur']
S4 = y['S4_cur']
S2 = y['S2_cur']
S1 = y['S1_cur']
Sp = y['Sp_cur']
V = y['V_cur']
EH = y['EH']
EL = y['EL']
```

```python
print(V)
```

```python
from sklearn.feature_selection import mutual_info_regression
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2


from sklearn.feature_selection import RFE
from sklearn.svm import SVR
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split


from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression


from mlxtend.feature_selection import ExhaustiveFeatureSelector as EFS
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

import seaborn as sns
```

# Voltage

### Embedded feature selection: Random Forest Importance

```python
X_V = X.copy()
X_V = X_V.drop('V_cur', axis = 1)
model = RandomForestRegressor()
model.fit(X_V, V)

feature_importances = pd.Series(model.feature_importances_, index=X_V.columns).sort_values(ascending=False)

# Plotting the feature importances
plt.figure(figsize=(9, 6))  # Adjust the size as needed
sns.barplot(x=feature_importances, y=feature_importances.index)

# Adding labels and title for clarity
plt.xlabel('Importance')
plt.ylabel('Features')
plt.title('Feature Importances from Random Forest')

plt.show()
```

# S8

### Filter method: Correction analysis

```python
X_S8_ca = X.copy()
X_S8_ca = X_S8_ca.drop('S8_cur', axis = 1)
X_S8_ca['S8'] = S8

correlation_matrix = X_S8_ca.corr()
target_correlation = correlation_matrix['S8'].sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=target_correlation.values, y=target_correlation.index)

plt.title('Correlation with S8 output')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Features')

plt.show()
```

### Filter method: Mutual information - non-linear

```python
# Assuming X, y are your features and target variable respectively
selector = SelectKBest(mutual_info_regression, k='all')  # Choose 'k' based on the number of features you want to select
X_selected = selector.fit_transform(X, S8)

print(X_selected)


# Calculate mutual information scores
mi_scores = mutual_info_regression(X, S8)

# Now, plot the scores
plt.figure(figsize=(12, 6))
plt.bar(range(X.shape[1]), mi_scores)
plt.xlabel('Features')
plt.ylabel('Mutual Information Score')
plt.title('Mutual Information Scores per Feature')

# Optional: If you want to have feature names and make it more readable
feature_names = ['S8_cur', 'S4_cur', 'S2_cur', 'S1_cur', 'V_cur', 'Sp_cur', 'EH', 'EL', 'iH', 'iL']  
plt.xticks(ticks=range(X.shape[1]), labels=feature_names, rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

### Embedded feature selection: Random Forest Importance

```python
X_S8_rf = X.copy()
X_S8_rf = X_S8_rf.drop('S8_cur', axis = 1)
model = RandomForestRegressor()
model.fit(X_S8_rf, S8)

feature_importances = pd.Series(model.feature_importances_, index=X_S8_rf.columns).sort_values(ascending=False)

# Plotting the feature importances
plt.figure(figsize=(9, 6))  # Adjust the size as needed
sns.barplot(x=feature_importances, y=feature_importances.index)

# Adding labels and title for clarity
plt.xlabel('Importance')
plt.ylabel('Features')
plt.title('Feature Importances from Random Forest')

plt.show()
```

### Wrapper method: Exhaustive Feature Selector

```python
X_efs = X.copy()

# Define your model
model = RandomForestRegressor()

# Initialize EFS with your model, specifying the scoring metric and the limits for the number of features
efs = EFS(model, 
          min_features=1, 
          max_features=6,  # Adjust based on your preference and computational constraints
          scoring='neg_mean_squared_error', 
          print_progress=True,
          cv=5)  # Cross-validation folds

# Fit EFS to your data
efs = efs.fit(X_efs, S8)

# Get the selected feature indices and print them
selected_features = efs.best_idx_
print(f'Selected features: {selected_features}')

# # You can then transform your dataset to keep only the selected features
# X_selected = efs.transform(X)
```
